```python
#  HW3

import pandas as pd
Location = "titanic3.xls"
Titanicdf = pd.read_excel(Location)
Titanicdf.head(18)
```

```python
#  On the whole women and girls were much more likely to survive than men and boys
#  Only about 1 in 5 males survived
Titanicdf.groupby('sex')['survived'].mean()
```

```python
#  The higher the class of room one was in, the better the chance of surviving
#  Only 1 in 4 passengers in third class survived
Titanicdf.groupby('pclass')['survived'].mean()
```

```python
#  Women/girls travelling in first and second class were the most fortunate passengers
#  Men/boys not travelling in first class were the least fortunate
Titanicdf.groupby(['sex','pclass'])['survived'].mean()
```

```python
#  Those who survived paid on average about twice as much as those who perished
Titanicdf.groupby('survived')['fare'].mean()
```

```python
#  Ratio is similar for the maximum fare paid
Titanicdf.groupby('survived')['fare'].max()
```

```python
#  Split up passengers into five age groups
agegrps = [0,13,20,40,60,100]
agenames = ['child','teen','yngadult','midage','senior']
Titanicdf['agegroup'] = pd.cut(Titanicdf['age'], agegrps, labels=agenames)
Titanicdf['agegroup'].unique()
```

```python
#  Middle-aged and older women were the most fortunate
#  Older men and teenage boys were the least fortunate
#  Younger boys had a little better than 50/50 chance of surviving

Titanicdf.groupby(['sex','agegroup'])['survived'].mean()
```

```python
#  These same percentages are reflected in this pivot table
pd.pivot_table(Titanicdf, values=['survived'], index=['agegroup'], columns=['sex'])
```

```python
import pandas as pd
Location = "algebradata.csv"
algebradf = pd.read_csv(Location)
algebradf.head()
```

```python
# HW1 part a)

TotStudent = len(algebradf)
Failed = len(algebradf[algebradf['Grade']=='F'])   # 'D' is considered a passing grade
Passed = TotStudent - Failed
PctPassed = float(Passed)/TotStudent * 100
PctPassed
```

```python
#  HW1 part b)

Womendf = algebradf[algebradf['Gender'] == 'F']
TotWomen = len(Womendf)
WomenFailed = len(Womendf[Womendf['Grade']=='F'])
WomenPassed = TotWomen - WomenFailed
PctPassed = float(WomenPassed)/TotWomen * 100
print(PctPassed)
```

```python
# HW1 part c)

algebradf['Hours of Study'].mean()
```

```python
# HW1 part d)

algebradf[algebradf['Grade']!='F']['Hours of Study'].mean()
```
